```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("titanic.csv")
df.head()
```

```python
df=df.dropna()
```

```python
inputs = df.drop(['PassengerId', 'Survived', 'Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'], axis='columns')
```

```python
target = df['Survived']
```

```python
#le_Pclass = LabelEncoder()
le_Sex = LabelEncoder()
# le_Age = LabelEncoder()
# le_Fare = LabelEncoder()
```

```python
#inputs['Pclass_n'] = le_Pclass.fit_transform(inputs['Pclass'])
inputs['le_Sex_n'] = le_Sex.fit_transform(inputs['Sex'])
# inputs['le_Age_n'] = le_Age.fit_transform(inputs['Age'])
# inputs['le_Fare_n'] = le_Fare.fit_transform(inputs['Fare'])
```

```python
inputs
```

```python
inputs_n = inputs.drop('Sex',axis='columns')
inputs_n
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(inputs_n,target,test_size=0.2)
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test,y_test)
```
